```python
import numpy as np
import tensorflow as tf
import pandas as pd
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from tensorflow.python.keras.callbacks import TensorBoard
from time import time


np.random.seed(3)
tf.random.set_seed(3)
```

```python
df = pd.read_csv('deeplearning/dataset/sonar.csv', header=None)
```

```python
dataset = df.values
X = dataset[:,0:60].astype(float)
Y_obj = dataset[:,60]

e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=3)
```

```python
model = Sequential()
model.add(Dense(24, input_dim=60, activation='relu')) #1464ê°œ
model.add(Dense(10, activation='relu')) #250
model.add(Dense(1, activation='sigmoid')) # 11
```

```python
tensorboard = TensorBoard(log_dir='logs\{}'.format(time()))
```

```python
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, Y_train, epochs=130, batch_size=5, callbacks=[tensorboard])
```

```python
model.save('my_model.h5')
del model
model = load_model('my_model.h5')
```

```python
print("\n Test Accuracy: %.4f" % (model.evaluate(X_test, Y_test)[1]))
```
